```python
import sys
import pickle
from os import mkdir
from os.path import exists, join
import re
from datetime import datetime, timedelta, date

import numpy as np
import pandas as pd
import scipy as sp

from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.decomposition import PCA
```

```python
feature_path = '../dataset/train/all_feature.pkl'
```

```python
not_use = [
    'orderid', 'uid', 'hotelid', 'basicroomid', 'hotel_roomid', 'roomid',
    'orderlabel'
]
```

```python
k = 400
```

```python
sample = None
```

```python
train_fatures = pd.read_pickle(feature_path)
print(train_fatures.shape)
print(train_fatures.columns)
train_columns = [x for x in train_fatures.columns if x not in not_use]
if len(train_columns) < k:
    k = len(train_columns)
select = SelectKBest(f_regression, k=k)
```

```python
train_f = train_fatures.sample(sample) if sample else train_fatures
```

```python
new_x = select.fit_transform(train_f[train_columns].fillna(0), train_f.orderlabel)
use_index = select.get_support()
```

```python
all_columns =  not_use + list(train_fatures[train_columns].loc[:, use_index].columns) 
```

```python
pickle.dump(all_columns, open('select_columns.pkl', 'wb'))
```
